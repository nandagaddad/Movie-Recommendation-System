```python
movies_data = pd.read_csv('/content/5000_movie_data_preprocessed.csv')
```

```python
movies_data.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer

# Create an instance of CountVectorizer
cv = CountVectorizer(max_features=5000, stop_words='english')

# Fit and transform the 'tags' column
vectors = cv.fit_transform(movies_data['tags']).toarray()

# Display the term-document matrix
print(vectors)

# Display the feature names (vocabulary)
print(cv.get_feature_names_out())
```

```python
vectors.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vectors)
```

```python
similarity[0]
```

```python
similarity.shape
```

```python
def recommend(movie):
    movie_index = movies_data[movies_data['title'] == movie].index[0]
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)),reverse = True,key=lambda x:x[1])[1:11]

    for i in movies_list:
        print(movies_data.iloc[i[0]].title)
```

```python
recommend('Avatar')
```
