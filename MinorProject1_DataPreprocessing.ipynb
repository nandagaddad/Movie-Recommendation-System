```python
movies = pd.read_csv('/content/tmdb_5000_movies.csv')
credits = pd.read_csv('/content/tmdb_5000_credits.csv')
```

```python
# now we are having two different data frames movies and credits so we will merge them

movies = movies.merge(credits,on='title')

movies.head(5000)    # in the output movie id,cast and crew are coming from the second data frames
```

```python
# columns which we are keeping
# 1. id
# 2. title
# 3. overview
# 4. genres
# 5. keywords
# 6. cast
# 7. crew
movies = movies[['movie_id' , 'title' , 'overview' , 'genres' , 'keywords' , 'cast' , 'crew']]
movies.head() # display the output
```

```python
movies.isnull().sum()
# we see that in overview we have 3 rows so we will drop them
```

```python
movies.dropna(inplace = True)
```

```python
movies.duplicated().sum()
```

```python

movies.iloc[0].genres # if we want to change the format of this then use the function below
```

```python
import ast
def convert(obj):
    L = []
    for i in ast.literal_eval(obj): # the obj that we get in output is a string after using ast.literal_eval it is converted into list
        L.append(i['name'])
    return L
```

```python
movies.head()
```

```python
movies['genres'] = movies['genres'].apply(convert)
movies['keywords'] = movies['keywords'].apply(convert)
movies.head()
```

```python
# we want only first 3 distionaries in cast section
import ast
def convert3(obj):
    L = []
    counter = 0
    for i in ast.literal_eval(obj): # the obj that we get in output is a string after using ast.literal_eval it is converted into list
        if counter != 3:
            L.append(i['name'])
            counter += 1
        else:
            break
    return L
```

```python
movies['cast'] = movies['cast'].apply(convert3)
```

```python
movies.head()
```

```python
import ast
def fetch_director(obj):
    L = []
    for i in ast.literal_eval(obj): # the obj that we get in output is a string after using ast.literal_eval it is converted into list
        if i['job'] == 'Director':
             L.append(i['name'])
             break
    return L
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
movies.head()
```

```python
movies['overview'][0]
```

```python
movies['overview'] = movies['overview'].astype(str).apply(lambda x:x.split())
```

```python
movies.head()
```

```python
# removing the space between actors and directors names, and keywords list elements
movies['genres'] = movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movies['cast'] = movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movies['crew'] = movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies.head()
```

```python
# we will make a new column called tags , in that we will Concatenate overview, genres, keywords, cast, crew all columns
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
movies.head()
```

```python
new_df = movies[['movie_id','title','tags']]
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x:" ".join(x))
```

```python
new_df.head()
```

```python
# for displaying particular column
new_df['tags'][0]
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x:x.lower())
new_df['tags'][0]
```

```python
new_df.head()
```

```python
from nltk.stem import PorterStemmer
# Initialize the stemmer
ps = PorterStemmer()

def stem(text):
    y = []

    for i in text.split():
        y.append(ps.stem(i))  # Apply stemming to each word

    return " ".join(y)  # Join the stemmed words back into a sentence

new_df['tags'] = new_df['tags'].apply(stem)
```
